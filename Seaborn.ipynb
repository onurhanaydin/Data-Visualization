```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
df=tips.groupby("day").sum()
df
```

```python
# with matplotlib

fig, ax = plt.subplots()
x = df.index
y = df.total_bill
ax.bar(x,y )
ax.set_title("Total Bill Distribution by Days")
for i in range(len(x)):
    plt.text(x[i], y[i], str(round(y[i],2)), ha="center",va="bottom")
```

# Pair Grid

```python
g=sns.PairGrid(tips)
g=g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(tips)

g=g.map_upper(sns.scatterplot) # yukarıları scatterplot yap
g=g.map_diag(sns.histplot)     # köşegenleri histplot yap
                               # aşağıyı tanımlamadığımız için boş bıraktı
```

```python
g=sns.PairGrid(tips)

g=g.map_upper(sns.scatterplot) 
g=g.map_diag(sns.histplot)
g=g.map_lower(sns.kdeplot, color="red")
```

```python
g = sns.PairGrid(tips, hue="sex", palette="viridis")
g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.distplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend();
```
