# Linewidth & Linestyles & Markers & Marker Size & Color

```python
x=np.arange(0,10)
x
```

# Linewidth

```python
fig, ax = plt.subplots(figsize=(12,6))
fig.suptitle("To change the line width, we can use the `linewidth` or `lw` keyword argument",fontsize=18)
ax.plot(x, x-1, color="red", linewidth=0.25)
ax.plot(x, x-2, color="red", lw=0.50)
ax.plot(x, x-3, color="red", lw=1)
ax.plot(x, x-4, color="red", lw=10);
```

# Linestyles

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
ax.plot(x, x-2, color="green", lw=3, ls='-.') # dash and dot
ax.plot(x, x-3, color="green", lw=3, ls=':') # dots
ax.plot(x, x-4, color="green", lw=3, ls='--') # dashes
plt.show()
```

# Markersize

```python
fig, ax = plt.subplots(figsize=(12,6))

# Use marker for string code
# Use markersize or ms for size

ax.plot(x, x-1,marker='+',markersize=20)
ax.plot(x, x-2,marker='o',ms=20) #ms can be used for markersize
ax.plot(x, x-3,marker='s',ms=20,lw=0) # make linewidth zero to see only markers
ax.plot(x, x-4,marker='1',ms=20);
```

```python
plt.figure(figsize=(15,5))
plt.suptitle("LineStyles & Markers & Markersize & Color",fontsize=16, y = 1.2)
plt.subplot(2,2,1)
plt.plot(age, salary, "r", ls="-", marker="+" ,markersize=30)
plt.subplot(2,2,2)
plt.plot(age, salary_2, "b",ls="-.", marker="o" ,markersize=30)
plt.subplot(2,2,3)
plt.plot(age, salary, "g",ls=":", marker="x" ,markersize=30)
plt.subplot(2,2,4)
plt.plot(age, salary_2, "y",ls="--" ,marker="*" ,markersize=30)
plt.tight_layout();
```

# Plot Range and Adding Extra Lines

```python
# Plot Range with set_ylim & set_xlim
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].set_xlim([1,3])
ax[1].set_ylim([1,50])
```

# Horizontal and Vertical Lines

```python
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[0].axvline(x=2,ls="--")
ax[0].axhline(y=20,ls="--")

ax[1].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].set_xlim([1,3])
ax[1].set_ylim([1,50])

ax[1].axvline(x=2,ls="--")
ax[1].axhline(y=20,ls="--");
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
```

```python
plt.figure(figsize=(10,8))
plt.pie(y, labels = mylabels, labeldistance = 0.7, autopct="%.2f"); 
# %.2f: "." 'dan sonra ka tane rakam almas gerektiini gsterir
# labeldistance yazlarn merkezden olan uzakln belirler
```

```python
plt.figure(figsize=(10,8))
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0, 0]

plt.figure(figsize=(10,8))
plt.pie(y, labels = mylabels, explode = myexplode, labeldistance=0.7, autopct="%.1f")
plt.legend();
```

```python
plt.figure(figsize=(10,8))

colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0, 0, 0.1, 0.1] # ratio for each continent with which to offset each wedge.

y = np.array([25, 35, 25, 15])

mylabels = ["Bananas", "Apples", "Cherries", "Dates"]

plt.pie(y, labels = mylabels,labeldistance=0.7, autopct="%.1f", startangle=90, shadow=True, pctdistance=1.1, colors=colors_list, explode=explode_list)
plt.legend();

# startangle: ilk label  90 dereceye denk getiriyor, daha sonrakileri de ona gre ayarlyor
```
